# Text Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(binary = True)
```

```python
corpus = ["NLP is awesome", "I want to start an NLP company", "Top unicorns use NLP"]
```

```python
vect.fit(corpus)
```

```python
vocab = vect.vocabulary_
```

```python
for key in vocab.keys():
    print ("{}: {}".format(key, vocab[key]))
```

```python
test_string_1 = "I have an awesome NLP Company worth an awesome billion"
print(vect.transform([test_string_1]), "\n\n")
print(vect.transform([test_string_1]).toarray())
```

## Mesuring The difference between 2 statements

```python
from sklearn.metrics.pairwise import cosine_similarity
sim = cosine_similarity(vect.transform(["ML is an elective subject now"]).toarray(), \
                        vect.transform(["NLP is an attractive career now"]).toarray())
print(sim)
```
